```python
import os, nltk, numpy as np, pandas as pd, jieba, jieba.analyse
from nltk import word_tokenize
from bs4 import BeautifulSoup

xue = pd.read_excel('C:\\Users\\fengzhi\\Desktop\\weibo_backup\\space_recycler.xls', index_col = None)
```

```python
tweet_raw = xue['content'].str.cat(sep='\n ')
soup = BeautifulSoup(tweet_raw, 'html.parser')
tweet_clean = soup.get_text()
```

```python
delete_words = ['微博', '转发','Repost','Share','pics','分享','图片','...','..', '哈哈哈','哈哈','哈哈哈哈','ll','啊啊啊','真的','网页','查看','链接','15','发现','哼哼','一个','喜欢','视频','朋友','全文']
for sw in delete_words:
    tweet_clean = tweet_clean.replace(sw, '')
```

```python
jieba.analyse.set_stop_words('C:\\Users\\fengzhi\\Desktop\\weibo_backup\\stopwords-master\\baidu_stopwords.txt')
jieba.analyse.extract_tags(tweet_clean, topK=50, withWeight=True, allowPOS=())
```

```python
jieba.analyse.textrank(tweet_clean, topK=50, withWeight=True, allowPOS=('ns', 'n', 'vn', 'v'))
```
